# Study Assignment: State Space Search
BFS and DFS on the 8-Puzzle problem.

```python
from collections import deque

GOAL = (1, 2, 3, 4, 5, 6, 7, 8, 0)
```

```python
def get_neighbors(state):
    state = list(state)
    i = state.index(0)
    row, col = i // 3, i % 3
    neighbors = []
    for dr, dc in [(-1,0),(1,0),(0,-1),(0,1)]:
        r, c = row + dr, col + dc
        if 0 <= r < 3 and 0 <= c < 3:
            j = r * 3 + c
            new = state[:]
            new[i], new[j] = new[j], new[i]
            neighbors.append(tuple(new))
    return neighbors
```

## BFS - Breadth First Search
Explores level by level. Always finds the shortest path.

```python
def bfs(start):
    queue = deque([[start]])
    visited = {start}
    while queue:
        path = queue.popleft()
        current = path[-1]
        if current == GOAL:
            return path
        for neighbor in get_neighbors(current):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(path + [neighbor])
    return None

start = (1, 2, 3, 4, 0, 6, 7, 5, 8)
path = bfs(start)
print("BFS solution in", len(path) - 1, "moves")
```

## DFS - Depth First Search
Explores as deep as possible first. Uses less memory but not always optimal.

```python
def dfs(start, limit=20):
    stack = [([start], {start})]
    while stack:
        path, visited = stack.pop()
        current = path[-1]
        if current == GOAL:
            return path
        if len(path) >= limit:
            continue
        for neighbor in get_neighbors(current):
            if neighbor not in visited:
                stack.append((path + [neighbor], visited | {neighbor}))
    return None

path = dfs(start)
print("DFS solution in", len(path) - 1, "moves")
```

## Comparing BFS and DFS on different starting states

```python
# Try a harder starting state
start2 = (1, 2, 3, 5, 0, 6, 4, 7, 8)

bfs_path = bfs(start2)
dfs_path = dfs(start2)

print("BFS moves:", len(bfs_path) - 1)
print("DFS moves:", len(dfs_path) - 1 if dfs_path else "Not found within limit")
```

## What happens when we change the DFS depth limit?

```python
for limit in [5, 10, 15, 20, 30]:
    path = dfs(start2, limit=limit)
    moves = len(path) - 1 if path else "Not found"
    print(f"Depth limit {limit}: {moves}")
```
