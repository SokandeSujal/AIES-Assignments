# Assignment 3: Constraint Satisfaction Problem
Solving SEND + MORE = MONEY using permutations.

```python
import itertools

def solve_csp():
    letters = ('S', 'E', 'N', 'D', 'M', 'O', 'R', 'Y')

    for p in itertools.permutations(range(10), len(letters)):
        s, e, n, d, m, o, r, y = p

        if s == 0 or m == 0:
            continue

        send  = s*1000 + e*100 + n*10 + d
        more  = m*1000 + o*100 + r*10 + e
        money = m*10000 + o*1000 + n*100 + e*10 + y

        if send + more == money:
            return dict(zip(letters, p))

solution = solve_csp()
print(solution)
```

## Verifying the solution

```python
s = solution

send  = s['S']*1000 + s['E']*100 + s['N']*10 + s['D']
more  = s['M']*1000 + s['O']*100 + s['R']*10 + s['E']
money = s['M']*10000 + s['O']*1000 + s['N']*100 + s['E']*10 + s['Y']

print(f"{send} + {more} = {money}")
print("Correct:", send + more == money)
```

## What if we try a different puzzle?
We can change the puzzle and apply the same approach.

```python
# BASE + BALL = GAMES
def solve_base_ball_games():
    letters = ('B', 'A', 'S', 'E', 'L', 'G', 'M')
    for p in itertools.permutations(range(10), len(letters)):
        b, a, s, e, l, g, m = p
        if b == 0 or g == 0:
            continue
        base  = b*1000 + a*100 + s*10 + e
        ball  = b*1000 + a*100 + l*10 + l
        games = g*10000 + a*1000 + m*100 + e*10 + s
        if base + ball == games:
            return dict(zip(letters, p)), base, ball, games
    return None, None, None, None

result, base, ball, games = solve_base_ball_games()
if result:
    print("Solution:", result)
    print(f"{base} + {ball} = {games}")
else:
    print("No solution found")
```

## N-Queens as a CSP

```python
def solve_n_queens(n):
    def is_safe(placement, row, col):
        for c, r in enumerate(placement):
            if r == row or abs(r - row) == abs(c - col):
                return False
        return True

    def backtrack(placement):
        col = len(placement)
        if col == n:
            return placement[:]
        for row in range(n):
            if is_safe(placement, row, col):
                placement.append(row)
                result = backtrack(placement)
                if result:
                    return result
                placement.pop()
        return None

    return backtrack([])

for n in [4, 6, 8]:
    sol = solve_n_queens(n)
    print(f"{n}-Queens:", sol)
```
